```python
stores = pd.read_csv(r"C:\Users\Marcelo\Documents\Hackaton\stores_details.csv",index_col=None, header=0)
print(stores.shape)
```

```python
full = pd.merge(combined_csv, stores, on='encrypted_cnpj')
```

```python
full.head()
full['productTotal'] = full.productTotal.astype(float)
full['productTotal'] = full['productTotal'].round(0)
full.drop("prod_fullname", inplace=True, axis=1)
full.drop("encrypted_domain_x", inplace=True, axis=1)
```

```python
full['totalCompra'] = full['productTotal'].groupby(full['encrypted_saleid']).transform('sum')
full.drop("productTotal", inplace=True, axis=1)
```

```python
zem = full['encrypted_buyer_cpf'].isna() & full['encrypted_buyer_cnpj'].isna()
pf = full['encrypted_buyer_cpf'].notnull()
pj = full['encrypted_buyer_cnpj'].notnull()

full['tipoComprador']=np.where(zem, 0,
                         np.where(pf, 1,
                         np.where(pj, 2, 0)))

full.drop("encrypted_buyer_cpf", inplace=True, axis=1)
full.drop("encrypted_buyer_cnpj", inplace=True, axis=1)
```

```python
full = full.drop_duplicates(subset='encrypted_saleid', keep="first")
```
