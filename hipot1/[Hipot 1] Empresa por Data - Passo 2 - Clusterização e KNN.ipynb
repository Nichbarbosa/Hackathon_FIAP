```python
# Importanto CSV de frequencias gerado pelo outro notebook

import pandas as pd

porcentagem_dias_transacoes_por_empresa = pd.read_csv("porcentagem_dias_transacoes_por_empresa.csv")

print(porcentagem_dias_transacoes_por_empresa.shape)
print(porcentagem_dias_transacoes_por_empresa.head(2))
```

```python
# Retirando empresas que tiveram < 100 transações

for index, row in porcentagem_dias_transacoes_por_empresa.iterrows():
    if row["Total"] < 100:
        porcentagem_dias_transacoes_por_empresa = porcentagem_dias_transacoes_por_empresa.drop(index)
        
print(porcentagem_dias_transacoes_por_empresa.shape)
print(porcentagem_dias_transacoes_por_empresa.head(2))
```

```python
# função para separar dataframe em test e training

import random

COLUNAS = ['dominio','Erros','Classe KMeans','Total']

dias = []

for data in range(1,32):
    dias.append(str(data))

COLUNAS =  COLUNAS + dias

trainingSet=pd.DataFrame(columns=COLUNAS)
testSet=pd.DataFrame(columns=COLUNAS)

def loadDataframes(dataframe, split, trainingSet , testSet):
    trainingIndex = 0
    testIndex = 0
    for index, row in dataframe.iterrows():
        #for y in range(23):
        #    row[str(y) + "h"] = float(row[str(y) + "h"])
        if random.random() < split:
            trainingSet.loc[trainingIndex] = dataframe.loc[index]
            trainingIndex += 1
        else:
            testSet.loc[testIndex] = dataframe.loc[index]
            testIndex += 1
```

```python
from sklearn.cluster import KMeans

X = porcentagem_dias_transacoes_por_empresa.iloc[:,3:34]

kmeans = KMeans(n_clusters=4, random_state=0).fit(X)
porcentagem_dias_transacoes_por_empresa['Classe KMeans'] = kmeans.fit_predict(X)

print(porcentagem_dias_transacoes_por_empresa.head(3))
```

```python
# Exportando arquivo Classes Kmeans

export_csv = porcentagem_dias_transacoes_por_empresa.to_csv (r'frequencia_dias_empresas_clusterizada.csv', index_label = "index")
```

```python
# Aplicando a função para separação de test e training

loadDataframes(porcentagem_dias_transacoes_por_empresa,0.67,trainingSet,testSet)

print(trainingSet.shape)
print(trainingSet.head(2))
print(testSet.shape)
print(testSet.head(2))
```

```python
# Calculo de distancias

import math

def euclideanDistance(instance1, instance2):
    distance = 0
    distance += pow((instance1["1"] - instance2["1"]), 2)
    distance += pow((instance1["2"] - instance2["2"]), 2)
    distance += pow((instance1["3"] - instance2["3"]), 2)
    distance += pow((instance1["4"] - instance2["4"]), 2)
    distance += pow((instance1["5"] - instance2["5"]), 2)
    distance += pow((instance1["6"] - instance2["6"]), 2)
    distance += pow((instance1["7"] - instance2["7"]), 2)
    distance += pow((instance1["8"] - instance2["8"]), 2)
    distance += pow((instance1["9"] - instance2["9"]), 2)
    distance += pow((instance1["10"] - instance2["10"]), 2)
    distance += pow((instance1["11"] - instance2["11"]), 2)
    distance += pow((instance1["12"] - instance2["12"]), 2)
    distance += pow((instance1["13"] - instance2["13"]), 2)
    distance += pow((instance1["14"] - instance2["14"]), 2)
    distance += pow((instance1["15"] - instance2["15"]), 2)
    distance += pow((instance1["16"] - instance2["16"]), 2)
    distance += pow((instance1["17"] - instance2["17"]), 2)
    distance += pow((instance1["18"] - instance2["18"]), 2)
    distance += pow((instance1["19"] - instance2["19"]), 2)
    distance += pow((instance1["20"] - instance2["20"]), 2)
    distance += pow((instance1["21"] - instance2["21"]), 2)
    distance += pow((instance1["22"] - instance2["22"]), 2)
    distance += pow((instance1["23"] - instance2["23"]), 2)
    distance += pow((instance1["24"] - instance2["24"]), 2)
    distance += pow((instance1["25"] - instance2["25"]), 2)
    distance += pow((instance1["26"] - instance2["26"]), 2)
    distance += pow((instance1["27"] - instance2["27"]), 2)
    distance += pow((instance1["28"] - instance2["28"]), 2)
    distance += pow((instance1["29"] - instance2["29"]), 2)
    distance += pow((instance1["30"] - instance2["30"]), 2)
    distance += pow((instance1["31"] - instance2["31"]), 2)
    return math.sqrt(distance)
```

```python
# Testando, se resposta for 0.034011 está ok (feito no excel)

print(euclideanDistance(porcentagem_dias_transacoes_por_empresa.loc[0],porcentagem_dias_transacoes_por_empresa.loc[1]))
```

```python
# Calculando os "vizinhos"

import operator 
def getNeighbors(trainingSet, testInstance, k):
    distances = []
    for x in range(len(trainingSet.index)):
        dist = euclideanDistance(testInstance, trainingSet.loc[x])
        distances.append((trainingSet.loc[x], dist))
    distances.sort(key=operator.itemgetter(1))
    neighbors = []
    for x in range(k):
        neighbors.append(distances[x][0])
    return neighbors
```

```python
# Testando vizinhos

neighbors = getNeighbors(trainingSet, testSet.loc[0], 2)
print(neighbors)
```

```python
# Predizendo a resposta 

import operator

def getResponse(neighbors):
    classVotes = {}
    for x in range(len(neighbors)):
        response = neighbors[x]["Classe KMeans"]
        if response in classVotes:
            classVotes[response] += 1
        else:
            classVotes[response] = 1
    sortedVotes = sorted(classVotes, reverse=True)
    return sortedVotes[0]
```

```python
# Pegando a acurácia do modelo

def getAccuracy(testSet, predictions):
    correct = 0
    for x in range(len(testSet.index)):
        if testSet.loc[x]["Classe KMeans"] is predictions[x]:
            correct += 1
    return (correct/float(len(testSet.index))) * 100.0
```

```python
# Testando o Modelo
predictions=[]

k = 5
    
for x in range(len(testSet.index)):
    neighbors = getNeighbors(trainingSet, testSet.loc[x], k)
    result = getResponse(neighbors)
    predictions.append(result)
    print('> predicted=' + repr(result) + ', actual=' + repr(testSet.loc[x]["Classe KMeans"]))

accuracy = getAccuracy(testSet, predictions)
print('Accuracy: ' + repr(accuracy) + '%')
```
